# Feature scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
```

# Support Vector Regression

```python
from sklearn.svm import SVR
```

```python
svr_reg = SVR(kernel='rbf')
```

```python
svr_reg.fit(X, Y)
```

```python
svr_pred = svr_reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("RMSE of SVR is", np.sqrt(mean_squared_error(Y, svr_pred)))
```

```python
svr_reg.score(X, Y)
```

# Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
ran_reg = RandomForestRegressor(n_estimators=200, random_state=0)
```

```python
ran_reg.fit(X, Y)
```

```python
ran_pred = ran_reg.predict(X)
```

```python
print("RMSE of Random forest is ", np.sqrt(mean_squared_error(Y, ran_pred)))
```

```python
ran_reg.score(X, Y)
```
